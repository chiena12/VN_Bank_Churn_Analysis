# NHÓM 2 : HÀNH VI VÀ ĐIỂM GẤP KHÚC 

## 1, Có phải càng sử dụng nhiều dịch vụ thì tỷ lệ exit càng giảm không ? Và từ bao nhiêu dịch vụ thì bắt đầu giảm mạnh ?

```python
WITH ServiceAnalysis AS (
    SELECT 
        nums_service,
        COUNT(*) AS total_customers,
        ROUND(SUM([exit]) * 100.0 /COUNT(*),2) AS exit_ratio
    FROM 
        vnbc
    GROUP BY nums_service
)

SELECT 
    nums_service,
    total_customers,
    exit_ratio,
    LAG(exit_ratio) OVER (ORDER BY nums_service) AS prev_exit_ratio,
    ROUND(LAG(exit_ratio) OVER (ORDER BY nums_service) - exit_ratio ,2) AS churn_reduction_impact 
FROM ServiceAnalysis
ORDER BY nums_service;
```

NHẬN XÉT : 

* Từ 1 lên 2 dịch vụ là bước làm quen hệ sinh thái và có mức giảm exit_ratio lớn nhất (7.93%)

* Thôg thường , mức giảm exit_ratio sẽ yếu dần theo số lượng dịch vụ .Tuy nhiên, ở mức từ 3 lên 4 dịch vụ thì mức giảm lại tăng từ 5,8% lên 6,5% .Điều này chứng tỏ ở mức 4 là mắt xích quan trọng , khiến cho khách hàng khó rời bỏ khách hàng nếu sử dụng đến mức này . Ngoài ra ,lượng khách hàng sử dụng 4 dịch vụ cũng khá đồng đều so với tệp khách hàng chỉ sử dụng 1 hay 2 dịch vụ 

==> Để thuyết phục khách hàng từ mức 4 lên mức 5 tốn nhiều chi phí về cả marketing và sales hơn nếu so với việc đưa khách hàng từ mức 1,2,3 lên mức 4 .Đây chính là vùng quan trọng nhất không chỉ mang lại lợi nhuận mà còn đem lại sự gắn kết bền vững nhất so với chi phí bỏ ra

## 2, Có phải tất cả các khách hàng đang sử dụng tích cực (active_member = 1) đều có engagement_score cao ? Tỷ lệ rời bỏ sẽ thay đổi như thế nào giữa các nhóm tuổi ?

```python
WITH 
    EngagementRanked AS (
        SELECT 
            age_group,
            active_member,
            [exit],
            engagement_score,
            CUME_DIST() OVER (ORDER BY engagement_score) AS engagement_pct    /* CUME_DIST dùng để tính phân vị */
        FROM vnbc 
    ),
    Classification AS (
        SELECT 
            CASE 
                WHEN active_member = 1 AND engagement_pct <= 0.5 THEN 'virtual_active_customer'
                WHEN active_member = 0 THEN 'truly_inactive'
                ELSE 'healthy_active' 
            END AS customer_classification,
            [exit],
            age_group
        FROM 
            EngagementRanked
    )

SELECT 
    customer_classification,
    age_group,
    COUNT(*) AS total_customer,
    ROUND(AVG(1.0*[exit])*100,2) AS exit_ratio
FROM 
    Classification
WHERE customer_classification IN ('virtual_active_customer','truly_inactive')
GROUP BY customer_classification,age_group

```

NHẬN XÉT : 

* Như vậy có thể thấy đa số khách hàng vẫn chưa phải là khách hàng tích cực cùng với đó là tỷ lệ exit khá đáng báo động

* Ở nhóm tuổi GenX chiếm số lượng khách hàng rất lớn nhưng tỷ lệ exit nếu inactive vẫn rất cao lên đến 20% trong khi vốn dĩ họ là nhóm khách hàng 'ổn định' và có tài chính vững nhất trong các nhóm tuổi .Điều này có thể do chuyển sang sử dụng dịch vụ các đối thủ cạnh tranh và điều này gây thất thoát lợi nhuận rất lớn

==> Cần có những chiến dịch 'Re-activate' đối với những khách hàng inactive - đặc biệt với GenX để biến họ thành những khách hàng tích cực vì chi phí bỏ ra sẽ thấp hơn nhiều so với tìm kiếm khách hàng mới .

## 3, Liệu độ trung thành(loyalty_level) cao và cách tương tác của khách hàng là online có làm tăng engagement_score hay không? Tỷ lệ exit như nào ?

## 1, Tổng tài sản mất mát (AUM)

```python
SELECT 
    customer_segment,
    COUNT(*) AS total_customer_exit,
    FORMAT(SUM(balance), 'N0') AS total_AUN_lost,
    FORMAT(AVG(balance), 'N0') AS avg_AUN_lost
FROM vnbc  
WHERE [exit] = 1
GROUP BY customer_segment
ORDER BY total_AUN_lost DESC
```

NHẬN XÉT : 

* Mặc dù về số lượng khách hàng thì phân khúc đại chúng vẫn tỷ trọng lớn nhất nhưng phân khúc khách hàng tiềm năng lại gây thất thoát của ngân hàng gấp 1,5 lần về cả tổng lẫn trung bình (157,8 tỷ so với 103,1 tỷ) .Bên cạnh đó ,mỗi khách hàng phân khúc này gây thất thoát của ngân hàng đến 28 triệu nếu rời bỏ.

* Giới thượng lưu và siêu giàu (Affluent && Priority) có số lượng ít hơn đáng kể nhưng chỉ cần 1 người rời cũng có thể khiến cho tài sản mất mát rất nhanh .Với phân khúc Priority thì tuy chỉ có 11 người nhưng mỗi người rời bỏ khiến cho ngân hàng mất đến 109,5 triệu - 1 con số rất lớn.

==> Ngân hàng nên có những chính sách ưu đãi đặc biệt nhằm dữ chân những khách hàng tiềm năng bởi họ là nhóm có khả năng rời đi chỉ sau nhóm đại chúng tuy nhiên lại là nhóm gây thất thoát nhiều nhất .Còn đối với phân khúc giàu có,cần có thêm những chương trình đặc biệt (Về lãi suất, đặc quyền,...) để kéo họ ở lại với ngân hàng

## 2, Tỷ lệ rời bỏ giữa thu nhập hàng tháng - nghề nghiệp 

```python
WITH 
    binned_income AS(   
        SELECT 
            occupation,
            [exit],
            CASE 
                WHEN monthly_ir < 15000000 THEN 'LOW'
                WHEN monthly_ir BETWEEN 15000000 AND 45000000 THEN 'MIDDLE'
                WHEN monthly_ir BETWEEN 45000000 AND 100000000 THEN 'HIGH'
                ELSE 'ELITE' 
            END AS income_segment
        FROM 
            vnbc
    )

SELECT 
    occupation AS [occupation\exit_ratio],
    ROUND(AVG(CASE WHEN income_segment = 'LOW' THEN CAST([exit] AS FLOAT) * 100.0 END),2) AS 'Low_Income_Exit_%',
    ROUND(AVG(CASE WHEN income_segment = 'MIDDLE' THEN CAST([exit] AS FLOAT) * 100.0 END),2) AS 'Middle_Income_Exit_%',
    ROUND(AVG(CASE WHEN income_segment = 'HIGH' THEN CAST([exit] AS FLOAT) * 100.0 END),2) AS 'High_Income_Exit_%',
    ROUND(AVG(CASE WHEN income_segment = 'ELITE' THEN CAST([exit] AS FLOAT) * 100.0 END),2) AS 'Elite_Income_Exit_%'
FROM 
    binned_income
GROUP BY occupation
ORDER BY occupation
```

NHẬN XÉT : 

* Đối với nhóm khách hàng Chủ Doanh Nghiệp và Quản lý/Lãnh đạo ,thu nhập của họ luôn từ Middle trở lên khi mà tỷ lệ Low_Income_Exit là giá trị NULL. Họ là phân khúc có thu nhập lớn hơn phần còn lại và một khi đang sử dụng 1 ngân hàng cố định với mức lương rất cao (Elite) thì họ chắc chắn không rời bỏ.Mặt khác thì Chủ Doanh nghiệp nhỏ ở mức thu nhập ở mức Middle thì có tỷ lệ rời lên đến 50% - điều này có thể giải thích với các lý do như rủi ro không tương xứng cũng như khát vốn nhưng lại khó tiếp cận  . Đối với các quản lý / lãnh đạo, họ có tỷ lệ exit chỉ là 2,4% nếu ở mức thu nhập Middle do họ không phải chịu rủi ro nếu như công ty làm ăn thua lỗ .Bên cạnh đó, thông thường các tài khoản ngân hàng là do cty chỉ định cũng như đặc thù công việc bận rộn nên họ không có nhiều thời gian suy nghĩ đổi ngân hàng

* Các kỹ sư IT có thu nhập hàng tháng cao hơn mặt bằng chung khi mà không có ai lương ở mức Low và tỷ lệ rời bỏ của họ cũng thấp ở mức 5,81%. Sở dĩ có điều này là do sự chọn lọc kỹ càng ngay từ ban đầu với những yêu cầu về tốc độ ,bảo mật, tính tự động ,... Họ cũng là phân khúc sẽ được hưởng những ưu đãi tốt từ ngân hàng.

* Các nhóm khách hàng như Giáo viên/Giảng viên , Kinh doanh/Bán hàng thì có tỷ lệ rời bỏ giảm dần theo thu nhập (Từ Low -> High) .Ở mức thu nhập thấp ,họ sẵn sàng rời bỏ để chuyển sang ngân hàng có chính sách ưu đãi hay các chương trình khuyến mãi tốt hơn. Mặt khác, khi tăng mức thu nhập lên Medium và High thì họ xu hướng dùng nhiều dịch vụ của ngân hàng hơn như gửi tiết kiệm hay vay tiêu dùng và gắn bó vơi 1 ngân hàng nhất định mà họ tin tưởng.

Họ dễ dàng chuyển qua ngân hàng khác khiến là ngân hàng đó đáp ứng được 1 tiêu chí nào đó mà họ đưa ra.
